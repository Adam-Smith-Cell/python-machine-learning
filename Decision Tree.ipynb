# Decision Tree & Random Forest With Python

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap
from sklearn.model_selection import train_test_split
from sklearn import tree, datasets
wine = datasets.load_wine()
x = wine.data[:,:2]
y = wine.target
x_train, x_test, y_train, y_test = train_test_split(x,y)
```

```python
clf = tree.DecisionTreeClassifier(max_depth=1)
clf.fit(x_train, y_train)
```

```python
cmap_light=ListedColormap(['#FFAAAA','#AAFFAA','#AAAAFF'])
cmap_bold=ListedColormap(['#FF0000','#00FF00','#0000FF'])
x_min, x_max = x_train[:,0].min()-1, x_train[:,0].max()+1
y_min, y_max = x_train[:,1].min()-1, x_train[:,1].max()+1
xx, yy = np.meshgrid(np.arange(x_min,x_max,0.02), np.arange(y_min,y_max,0.02))
z = clf.predict(np.c_[xx.ravel(), yy.ravel()])
z = z.reshape(xx.shape)
plt.figure()
plt.pcolormesh(xx,yy,z,cmap=cmap_light)
plt.scatter(x[:,0], x[:,1], c=y, cmap=cmap_bold, edgecolor='k',s=20)
plt.xlim(xx.min(),xx.max())
plt.ylim(yy.min(),yy.max())
plt.title("Classifier:(max_depth=1)")
plt.show()
```

```python
clf2 = tree.DecisionTreeClassifier(max_depth=3)
clf2.fit(x_train, y_train)
```

```python
cmap_light=ListedColormap(['#FFAAAA','#AAFFAA','#AAAAFF'])
cmap_bold=ListedColormap(['#FF0000','#00FF00','#0000FF'])
x_min, x_max = x_train[:,0].min()-1, x_train[:,0].max()+1
y_min, y_max = x_train[:,1].min()-1, x_train[:,1].max()+1
xx, yy = np.meshgrid(np.arange(x_min,x_max,0.02), np.arange(y_min,y_max,0.02))
z = clf2.predict(np.c_[xx.ravel(), yy.ravel()])
z = z.reshape(xx.shape)
plt.figure()
plt.pcolormesh(xx,yy,z,cmap=cmap_light)
plt.scatter(x[:,0], x[:,1], c=y, cmap=cmap_bold, edgecolor='k',s=20)
plt.xlim(xx.min(),xx.max())
plt.ylim(yy.min(),yy.max())
plt.title("Classifier:(max_depth=3)")
plt.show()
```

```python
clf3 = tree.DecisionTreeClassifier(max_depth=5)
clf3.fit(x_train, y_train)
```

```python
z = clf3.predict(np.c_[xx.ravel(), yy.ravel()])
z = z.reshape(xx.shape)
plt.figure()
plt.pcolormesh(xx,yy,z,cmap=cmap_light)
plt.scatter(x[:,0], x[:,1], c=y, cmap=cmap_bold, edgecolor='k',s=20)
plt.xlim(xx.min(),xx.max())
plt.ylim(yy.min(),yy.max())
plt.title("Classifier:(max_depth=5)")
plt.show()
```

```python
from sklearn.ensemble import RandomForestClassifier
forest = RandomForestClassifier(n_estimators=6, random_state=3)
forest.fit(x_train, y_train)
```

```python
z = forest.predict(np.c_[xx.ravel(), yy.ravel()])
z=z.reshape(xx.shape)
plt.figure()
plt.pcolormesh(xx,yy,z,cmap=cmap_light)
plt.scatter(x[:,0], x[:,1], c=y, cmap=cmap_bold, edgecolor='k',s=20)
plt.xlim(xx.min(),xx.max())
plt.ylim(yy.min(),yy.max())
plt.title("Classifier:(max_depth=5)")
plt.show()
```

```python
forest.score(x_test, y_test)
```

## 决策树&随机森林 实战

```python
import pandas as pd
data = pd.read_csv('adult.csv',header=None,index_col=False,names=['年龄','单位性质','权重','学历','受教育时长','婚姻状况','职业','家庭状况','种族','性别','资产所得','资产损失','周工作时长','国籍','收入'])
data_lite = data[['年龄','单位性质','学历','性别','周工作时长','职业','收入']]
display(data_lite.head())
```

```python
data_dummies = pd.get_dummies(data_lite)
list(data_lite.columns), list(data_dummies.columns)
```

```python
data_dummies.head()
```

```python
features = data_dummies.iloc[:,:-2]
x = features.values
y = data_dummies['收入_ >50K'].values
```

```python
x.shape, y.shape
```

```python
x_train, x_test,y_train,y_test=train_test_split(x,y,random_state=0)
go_dating_tree = tree.DecisionTreeClassifier(max_depth=5)
go_dating_tree.fit(x_train, y_train)
go_dating_tree.score(x_test, y_test)
```
